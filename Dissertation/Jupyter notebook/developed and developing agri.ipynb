```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib notebook


df=pd.read_excel('agri comparison.xlsx')
   
# plotting three histograms on the same axis
plt.hist(df['Australia'],bins = 5, alpha = 0.5, 
         color = 'green',edgecolor='k')
plt.hist(df['United Kingdom'],bins = 5, alpha = 1,
         color = 'blue',edgecolor='k')
plt.hist(df['Nepal'],bins = 30, alpha = 0.5,
         color = 'red',edgecolor='k')
plt.hist(df['Philippines'],bins = 30, alpha = 0.5, 
         color = 'orange',edgecolor='k')


plt.title("Agriculture, forestry, and fishing(AFF), value added in GDP")
plt.legend(['Australia','United Kingdom','Nepal','Philippines'])
plt.xlabel('AFF value % from 1990-2022')
plt.ylabel('Total Number of same AFF value occuring from 1990-2022')
plt.show()
```
