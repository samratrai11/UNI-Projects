```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib notebook


df=pd.read_excel('industry all.xlsx')
   
# plotting three histograms on the same axis
plt.hist(df['Australia'],bins = 5, alpha = 0.5, 
         color = 'green',edgecolor='k')
plt.hist(df['United Kingdom'],bins = 5, alpha = 1,
         color = 'blue',edgecolor='k')
plt.hist(df['Nepal'],bins = 30, alpha = 0.5,
         color = 'silver',edgecolor='k')
plt.hist(df['Philippines'],bins = 30, alpha = 0.5, 
         color = 'yellow',edgecolor='k')


plt.title("Industry (including construction), value added (% of GDP)")
plt.legend(['Australia','United Kingdom','Nepal','Philippines'])
plt.xlabel('Industry value % from 1990-2022')
plt.ylabel('Total Number of industry value occuring from 1990-2022')
plt.show()
```
